```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
sns.set(style="darkgrid")
```

```python
df = pd.read_excel('data/raw/jesterfinal151cols.xls', header = None)
display(df)
```

```python
# remove never rated jokes
removed = [1, 2, 3, 4, 5, 6, 9, 10, 11, 12, 14, 20, 27, 31, 43, 51, 52, 61, 73, 80, 100, 116]
df.drop(df.columns[removed], axis=1, inplace=True)
display(df)
```

```python
# Number of nonzero entries
counts = df[0].value_counts()
sns.relplot(data=counts)
plt.xlabel('Number of jokes rated')
plt.ylabel('Number of users')
```

```python
# Clean up jokes
import re
import csv
from bs4 import BeautifulSoup

def remove_html_tags(raw_html):
    pat = re.compile('<.*?>')
    clean_html = re.sub(pat, '', raw_html)
    return clean_html

def clean_jokes(data_file):
    with open(data_file, 'r') as f:
        html_doc = f.read()

    soup = BeautifulSoup(html_doc, 'html.parser')
    paras = soup.find_all('p')
    docs = [remove_html_tags(p.text) for p in paras]
    docs = [re.sub('\s+', ' ', doc) for doc in docs]

    with open('data/jokes/clean_jokes.tsv', 'w') as out:
        for i, doc in enumerate(docs):
            out.write(str(i) + '\t' + doc + '\n')

data_file = 'data/jokes/jester_items.dat'
clean_jokes(data_file)
```

```python
# Some stats about the jokes
jokes = pd.read_table('data/jokes/clean_jokes.tsv', sep='\t', header = None)

# replace 99 with NaN
df = df.replace(99, np.NaN)

# mean joke rating
ave_jokes = df.iloc[:,1:].aggregate(axis=0, func='mean').sort_values()

# funniest joke on average
print('joke id:', ave_jokes.idxmax(), 'score:', ave_jokes.max())
print('joke:', jokes.iloc[ave_jokes.idxmax()][1], '\n')
# least funny joke
print('joke id:', ave_jokes.idxmin(), 'score:', ave_jokes.min())
print('joke:', jokes.iloc[ave_jokes.idxmin()][1])


```

```python
joke_values = ave_jokes.dropna()
sns.distplot(joke_values)
plt.xlabel('Average score')
plt.ylabel('Ratio of jakes')
plt.title('Distribution of joke scores')
```
